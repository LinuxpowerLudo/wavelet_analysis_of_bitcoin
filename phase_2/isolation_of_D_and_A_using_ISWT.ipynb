```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16, 3)

import pywt
hor_len=12
vert_len=3
```

```python
dataset = pd.read_csv("../datasets/ibm_returns.csv", header=None, index_col=0)
```

```python
abs_log_returns = np.abs(dataset[dataset.columns[0]])
```

```python
v_n = abs_log_returns.values
plt.plot(abs_log_returns)
plt.title("original_series")
plt.show()
num_decompositions = 4
data_transformed = pywt.swt(v_n, "haar", level=4)
for i in range(num_decompositions, 0, -1):
    level = num_decompositions+1-i
    a, d = data_transformed[i-1]
    d = pywt.iswt((np.zeros(d.shape), d), 'haar')
    for j in range(level-1):
        d = iswt((d, np.zeros(d.shape)))
    
    plt.title("details "+str(num_decompositions+1-i))
    plt.plot(d)
    plt.show()
    
    if i == 1:
    
        plt.title("smooth "+str(num_decompositions+1-i))
        plt.plot(a)
        plt.show()
    '''
    figure(num=None, figsize=(hor_len, vert_len), dpi=180, facecolor='w', edgecolor='k')

    plt.title("smooth "+str(4-j))
    plt.plot(a)
    plt.show()'''
```
