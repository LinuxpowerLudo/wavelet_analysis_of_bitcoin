```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure


import pywt
```

```python

dataset = pd.read_csv("ibm_returns.csv", header=None, index_col=0)
abs_log_returns = np.abs(dataset[dataset.columns[0]])
series_length = abs_log_returns.shape[0]
```
